```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("owid-covid-data.csv")
```

```python
#ANALISI DELLE COLONNE NUMERICHE
df.describe()
```

```python
#ANALISI DELLE COLONNE CATEGORICHE
for column in df.columns:
    if df[column].dtype == "object":
        print(f"{column} column contains categorical data.")
        unique_count = df[column].nunique()
        print(f"{column} contains {unique_count} unique values")
        all_count = df[column].value_counts()
        print(f"Distribution of every unique value is as such:")
        print(all_count)
        print()
```

```python
#INDIVIDUAZIONE DEI DATI NULLI
for column in df.columns:
    nan_count = df[column].isna().sum()
    nan_percentage = round((nan_count/df.shape[0])*100,2)
    print(f"{column} contains {nan_count} NaN values, {nan_percentage}% of all rows.")
```

```python
#per ogni continente il numero di casi fin dall'inizio della pandemia
casi_continente = df.groupby("continent")["new_cases"].sum()
print(casi_continente)
```

```python
#per ogni continente la percentuale rispetto al totale mondiale del numero di casi
casi_mondo = casi_continente.sum()
print(f"Il numero totale di casi nel mondo è:", casi_mondo)
```

```python
percentuale_casi_continente = ((casi_continente/casi_mondo)*100).round(2)
print(percentuale_casi_continente)
```

```python
#Selezionare i dati relativi all'Italia nel 2022
df['date'] = pd.to_datetime(df['date'])
italia_2022 = df[(df['location'] == 'Italy') & (df['date'].dt.year == 2022)]
display(italia_2022)
```

```python
#filtrare via i giorni che non hanno misurazioni, quindi filtro per i giorni con valore diverso da 0.0
italia_2022 = italia_2022[italia_2022['new_cases']!= 0.0]
display(italia_2022)
```

```python
#l'evoluzione del casi totali dall'inizio alla fine dell'anno
italia_2022["casi_totali_milioni"] = italia_2022["total_cases"]/1_000_000
italia_2022.plot(x="date", y="casi_totali_milioni", kind="line", title="Evoluzione casi totali in Italia 2022")
plt.xlabel("Data")
plt.ylabel("Casi Totali (in milioni)")
plt.grid()
plt.show()
```

```python
#il numero di nuovi casi rispetto alla data
italia_2022.plot(x="date", y="new_cases", kind="line", title="Nuovi casi giornalieri in Italia 2022")
plt.xlabel("Data")
plt.ylabel("Nuovi casi (milioni)")
plt.grid()
plt.show()
```

```python
#nazioni Italia, Germania, Francia
#boxplot: differenza n pazienti in terapia intensiva (icu_patiens), maggio 2022-aprile 2023
it_ger_fr_df = df[
    (df['location'].isin(['Italy', 'Germany', 'France'])) &
    (df['date'] >= "2022-05-01") &
    (df['date'] <= "2023-04-30")][['iso_code','continent','location','date','total_cases','new_cases','icu_patients']]
display(it_ger_fr_df)
```

```python
it_ger_fr_df.loc[it_ger_fr_df['icu_patients'].isna()]
```

```python
#prima di gestire i dati nulli, guardo se ci sono altri valori nel mese successivo
df[(df['location'] == 'France') & 
(df['date'] >= '2023-05-01') &
(df['date'] <= '2023-05-31')][['date', 'icu_patients']]
```

```python
#elimino i valori nulli 
it_ger_fr_df = it_ger_fr_df.dropna(subset=['icu_patients'])
it_ger_fr_df
```

```python
#boxplot
sns.boxplot(data=it_ger_fr_df, x="location", y="icu_patients")
```

```python
# Guardando il grafico (boxplot), possiamo notare alcune differenze tra i tre paesi.  
# Tuttavia, è importante considerare che la Francia ha un mese in meno in analisi (Aprile 2023),  
# poiché i dati per quel periodo non sono disponibili.  
# 
# Di conseguenza, il confronto diretto tra i paesi deve tenere conto di questa differenza.  
# Possiamo comunque osservare che in Francia e Germania, durante il periodo considerato,  
# c'è stata una maggiore pressione sul sistema sanitario.  
```

```python
#Riguardo le nazioni di Italia, Germania, Francia e Spagna in tutto il 2021:
it_ger_fr_spa_df = df[
    (df['location'].isin(['Italy', 'Germany', 'France','Spain'])) &
    (df['date'] >= "2021-01-01") &
    (df['date'] <= "2021-12-31")][['iso_code','continent','location','date','total_cases','new_cases','hosp_patients']]
display(it_ger_fr_spa_df)
```

```python
#vedo se ci sono i valori nulli
it_ger_fr_spa_df.loc[it_ger_fr_spa_df['hosp_patients'].isna()]
```

```python
#ci sono valori nulli per tutto il 2021 per la Germania
#dunque, guardo se nel dataframe ci sono altri valori, per capire se posso fare una sostituzione o meno
df[(df['location'] == 'Germany') & 
(df['date'] >= '2020-12-31') &
(df['date'] <= '2022-02-01')][['location','date', 'hosp_patients']]
```

```python
#sono tutti valori nulli
#quindi elimino i valori nulli 
it_ger_fr_spa_df= it_ger_fr_spa_df.dropna(subset=['hosp_patients'])
it_ger_fr_spa_df
```

```python
#faccio la somma per ogni paese dei pazienti ospedalizzati
sum_hosppa_country = it_ger_fr_spa_df.groupby('location')['hosp_patients'].sum()
sum_hosppa_country 
```

```python
#commento
#come verificato precedentemente la Germania ha tanti valori nulli, concentrati in tutto il 2021, anche nei mesi successivi. 
#Per questo motivo è stato preferito rimuoverli senza effettuare alcuna sostituzione. 
#Nel grafico non vedrete la Germania.
```

```python
#grafico
sum_hosppa_country.plot(kind="bar", title="Totale pazienti ospedalizzati nel 2021 per nazione",color=['purple', 'yellow', 'green'])
plt.xlabel("Paese")
plt.ylabel("Pazienti ospedalizzati (milioni)")
plt.grid()
plt.show()
```
